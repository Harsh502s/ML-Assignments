## Barplot of SepalLength vs Species

```python
plot1df = df[["SepalLengthCm", "Species"]].groupby("Species").median()
plot1df.reset_index(inplace=True)

sns.barplot(x="Species", y="SepalLengthCm", data=plot1df)
```

### From the plot we got to know that Sepal length is in the Setosa<Versicolor<Virginica.

## Barplot of SepalWidth vs Species

```python
plot2df = df[["SepalWidthCm", "Species"]].groupby("Species").median()
plot2df.reset_index(inplace=True)

sns.barplot(x="Species", y="SepalWidthCm", data=plot2df)
```

### From the plot we got to know that Sepal Width of all the category of Iris are similar.

## Barplot of PetalLength vs Species

```python
plot3df = df[["PetalLengthCm", "Species"]].groupby("Species").median()
plot3df.reset_index(inplace=True)

sns.barplot(x="Species", y="PetalLengthCm", data=plot3df)
```

### From the plot we got to know that Petal length of Virginica is way much more than Versicolor and the lest Petal length is of Setosa.

## Barplot of PetalWidth vs Species

```python
plot4df = df[["PetalWidthCm", "Species"]].groupby("Species").median()
plot4df.reset_index(inplace=True)

sns.barplot(x="Species", y="PetalWidthCm", data=plot4df)
```

### From the plot we got to know that Petal Width of Virginica is way much more than Versicolor and the smallest Petal length is of Setosa.
